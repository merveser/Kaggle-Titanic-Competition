**1.Find the average age of those who died in the accident**

```python
avg_of_dead = df[df['Survived'] == 0]['Age'].mean()
print(f"Average age of those who died in the accident: {avg_of_dead:.2f}")
```

**2.Find the mean and median ticket prices of those who died in the accident**

```python
avg_fare = df[df['Survived'] == 0]['Fare'].mean()
median_fare = df[df['Survived'] == 0]['Fare'].median()
print(f"Average fare of those who died in the accident: {avg_fare:.2f}")
print(f"Median fare of those who died in the accident: {median_fare:.2f}")
```

**3.Find the average age of the men who died in the accident**

```python
avg_age_dead_men = df[(df['Survived'] == 0) & (df['Sex'] == 'male')]['Age'].mean()
print(f"The average age of men who died in an accident: {avg_age_dead_men:.2f}")
```

**4.Find the average age of women who died in the accident**

```python
avg_age_dead_women = df[(df['Survived'] == 0) & (df['Sex'] == 'female')]['Age'].mean()
print(f"The average age of women who died in an accident: {avg_age_dead_women:.2f}")
```

```python
survived = df[df['Survived'] == 1]
survived[['Name', 'Age']]
```

**5.Find the average age of those who survived the accident**

```python
avg_of_survived = df[df['Survived'] == 1]['Age'].mean()
print(f"Average age of those who survived in the accident: {avg_of_survived:.2f}")
```

**6.Find the average ticket price of those who survived the accident**

```python
avg_fare_survived = df[df['Survived'] == 1]['Fare'].mean()
print(f"Average fare of those who survived in the accident: {avg_fare_survived:.2f}")
```

**7.Find the total number of people who survived the accident**

```python
survivors = len(df[df['Survived'] == 1])
print(f"Total number of people who survived the accident: {survivors}")
```

**8.Find the median value of ticket prices for those under 10 years old**

```python
avg_fare_children = df[df['Age'] < 10]['Fare'].median()
print(f"Median ticket prices for children under 10 years old: {avg_fare_children:.2f}")
```


**9.Compare the average and median of 1st class, 2nd class and 3rd class ticket prices.**

```python
Pclass_mean1 = df[df['Pclass'] == 1]['Fare'].mean()
Pclass_mean2 = df[df['Pclass'] == 2]['Fare'].mean()
Pclass_mean3 = df[df['Pclass'] == 3]['Fare'].mean()
print(f"Average of 1st class ticket price: {Pclass_mean1:.2f}\n"
      f"Average of 2nd class ticket price: {Pclass_mean2:.2f}\n"
      f"Average of 3rd class ticket price: {Pclass_mean3:.2f}")
```

```python
Pclass_median1 = df[df['Pclass'] == 1]['Fare'].median()
Pclass_median2 = df[df['Pclass'] == 2]['Fare'].median()
Pclass_median3 = df[df['Pclass'] == 3]['Fare'].median()

print(f"Median of 1st class ticket price: {Pclass_median1:.2f}\n"
      f"Median of 2nd class ticket price: {Pclass_median2:.2f}\n"
      f"Median of 3rd class ticket price: {Pclass_median3:.2f}")
```

```python
for pclass in sorted(df['Pclass'].unique()):
    avg = df[df['Pclass'] == pclass]['Fare'].mean()
    med = df[df['Pclass'] == pclass]['Fare'].median()
    print(f"{pclass}. class - Average Fare: {avg:.2f}, Median Fare: {med:.2f}")
```

**10.Compare the ratio of women who died in the accident to the ratio of men**

```python
dead_women_ratio = len(df[(df['Survived'] == 0) & (df['Sex'] == 'female')]) / len(df[df['Sex'] == 'female'])
dead_men_ratio = len(df[(df['Survived'] == 0) & (df['Sex'] == 'male')]) / len(df[df['Sex'] == 'male'])

if dead_women_ratio > dead_men_ratio:
    print(f"A higher percentage of women died in the accident compared to men.\n"
          f"Women: {dead_women_ratio:.2%}, Men: {dead_men_ratio:.2%}")
else:
    print(f"A higher percentage of men died in the accident compared to women.\n"
          f"Men: {dead_men_ratio:.2%}, Women: {dead_women_ratio:.2%}")
```
